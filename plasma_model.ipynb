```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate dummy data
np.random.seed(0)
voltage = np.random.uniform(1000, 3000, 50)  # Random voltage between 1000V and 3000V
temperature = np.random.uniform(30, 50, 50)  # Random temperature between 30°C and 50°C
flow_speed = 0.003 * voltage + 0.02 * temperature + np.random.normal(0, 0.5, 50)  # Adding noise

# Create a DataFrame
data = pd.DataFrame({
    'Voltage (V)': voltage,
    'Temperature (°C)': temperature,
    'Flow Speed (m/s)': flow_speed
})

# Check the first 5 rows of the data
print(data.head())

# Features (Voltage, Temperature) and target (Flow Speed)
X = data[['Voltage (V)', 'Temperature (°C)']]
y = data['Flow Speed (m/s)']

# Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# 3D Plot for Visualization
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Voltage (V)'], data['Temperature (°C)'], y, color='blue', label='Actual Data')
ax.set_xlabel('Voltage (V)')
ax.set_ylabel('Temperature (°C)')
ax.set_zlabel('Flow Speed (m/s)')
plt.legend()
plt.show()

# Display the regression coefficients and intercept
print("Regression coefficients (slope):", model.coef_)
print("Intercept:", model.intercept_)

# Predicting Flow Speed for new data (e.g., Voltage 2500V, Temperature 40°C)
new_data = np.array([[2500, 40]])  # New data (Voltage, Temperature)
predicted_flow = model.predict(new_data)
print("Predicted Flow Speed (m/s):", predicted_flow)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from typing import Union

# ダミーデータを作成（既存のデータを使用）
data = {
    "voltage": [2000, 2200, 2400, 2600, 2800],
    "temperature": [30, 32, 34, 36, 38],
    "flow_speed": [5, 5.5, 6, 6.5, 7],
}

# データをPandas DataFrameに変換
df = pd.DataFrame(data)

# 線形回帰モデルの作成
X = df[["voltage", "temperature"]]
y = df["flow_speed"]

model = LinearRegression()
model.fit(X, y)

# FastAPIインスタンスを作成
app = FastAPI()

# 入力データのスキーマを定義
class InputData(BaseModel):
    voltage: float
    temperature: float

# 流速を予測するエンドポイント
@app.post("/predict/")
def predict_flow_speed(input_data: InputData):
    voltage_input = input_data.voltage
    temperature_input = input_data.temperature
    predicted_flow = model.predict(np.array([[voltage_input, temperature_input]]))
    return {"predicted_flow_speed": predicted_flow[0]}

```

```python
import os
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import random
from fastapi import FastAPI

app = FastAPI()

@app.get("/save_data_to_csv/")
def save_data_to_csv():
    # ダミーデータを生成
    np.random.seed(0)
    voltage = np.random.uniform(1000, 3000, 50)  # ランダムな電圧
    temperature = np.random.uniform(30, 50, 50)  # ランダムな温度
    flow_speed = 0.003 * voltage + 0.02 * temperature + np.random.normal(0, 0.5, 50)  # 流速
    
    # 日付を生成（過去30日間のランダムな日付）
    today = datetime.today()
    date_range = [today - timedelta(days=random.randint(0, 30)) for _ in range(50)]
    dates = [date.strftime('%Y-%m-%d') for date in date_range]
    
    # データをDataFrameに格納
    data = pd.DataFrame({
        'Date': dates,  # 日付を追加
        'Voltage (V)': voltage,
        'Temperature (°C)': temperature,
        'Flow Speed (m/s)': flow_speed
    })

    # CSVファイルに保存
    file_path = "dummy_data_with_date.csv"
    data.to_csv(file_path, index=False)

    return {"message": f"Data saved to {file_path}"}

# save_data_to_csv()を手動で呼び出す場合：
save_data_to_csv()
```
